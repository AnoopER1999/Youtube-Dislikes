## 13). Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

```python
new = DF1[['view_count','dislikes']]
```

### Co-orelation Matrix

```python
new.corr()
```

### Covariance Matrix

```python
new.cov()
```

```python
plt.figure(figsize=(7,5))
plt.ticklabel_format(style='sci')
plt.scatter(new['view_count'],new['dislikes'])
plt.title('View_count v/s Dislikes')
plt.xlabel('view_count')
plt.ylabel('dislikes')
plt.style.use(style='ggplot')
plt.show()
```

# Inference

- From the co-orelation and covariance matrix we can say that they have a positive relation between each other. i.e, if one increases other also increases and vice versa.
- Scatter plot also implies the same.
